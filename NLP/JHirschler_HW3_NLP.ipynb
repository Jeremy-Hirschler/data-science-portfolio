```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
import nltk
nltk.download('punkt')

from nltk import ngrams
from nltk.tokenize import word_tokenize
```

### 1. Read the Corpus

```python
with open('./corpus.en', 'r', encoding='utf-8') as file:
    text = file.read()
print(text[:100])
```

We see we only have a single column here with various phrases and text. Let's move on to preprocessing.

### 2. Preprocess the Text

```python


def preprocess_text(text):
    # Remove symbols and keep words/spaces
    text = re.sub(r'[^\w\s]', '', text)

    # Convert to lowercase
    text = text.lower()

    # Tokenize using NLTK
    tokens = word_tokenize(text)

    return text, tokens
```

```python
cleaned_text, tokens = preprocess_text(text)
# print(f'Cleaned Text: {cleaned_text}')
print(f'Tokens: {tokens}')
```

We can see the corpus have been divided into tokens for each word. We used regex to remove the symbols and punctuation marks and the word tokenize function from the nltk library.

### 3. Generate Trigrams

Let's create trigrams also with the help of nltk library.

```python
# generate ngrams for any number n
def get_ngrams(tokens, n):

    # Generate ngrams
    ngrams_list = list(ngrams(tokens, n))
    return ngrams_list
```

```python
trigrams = get_ngrams(tokens, 3)
print(f'Trigrams: {trigrams}')
```

We can also calculate the bigrams and unigrams for the corpus to use in the next few sections.

```python
bigrams = get_ngrams(tokens, 2)
unigrams = get_ngrams(tokens, 1)
```

We can see we have trigrams where the sets are 3 words long each, and each subsequent trigram includes the the last 2 words from the previous.

### Calculate Trigram Frequencies

Now let's calculate the trigram frequencies.

```python
from collections import Counter
```

```python
def get_ngram_frequencies(ngrams_list):
    # Count ngram frequencies
    ngram_freq = Counter(ngrams_list)
    
    return ngram_freq
```

```python
trigram_freq = get_ngram_frequencies(trigrams)
bigram_freq = get_ngram_frequencies(bigrams)
unigram_freq = get_ngram_frequencies(unigrams)
```

```python
# Display some trigram and bigram frequencies
print("Trigram frequencies:", trigram_freq.most_common(5))
print("Bigram frequencies:", bigram_freq.most_common(5))
```

We can see the top 5 frequencies of both the trigrams and bigrams. Of course the bigrams will naturally have more occurrences than the trigrams due to matching fewer tokens.

### Compute Conditional Probabilities

Now we can compute the conditional probabilities.

```python
def compute_trigram_probability(w1, w2, w3):
    
    # Get counts for the specific trigrams and bigrams
    trigram_count = trigram_freq[(w1, w2, w3)]
    bigram_count = bigram_freq[(w1, w2)]

    # Calculate probability
    if bigram_count > 0:
        probability = trigram_count / bigram_count
    else:
        probability = 0  # Avoid division by zero if bigram never occurs

    return probability
```

### Handle Unknown Words

We can use a stupid backoff mechanism that will try for n-1 ngrams if the trigram and bigram isn't found.

```python
def backoff_probability(w1, w2, w3, alpha=0.4):
    '''Computes probability using a backoff mechanism.'''
    total_unigrams = sum(unigram_freq.values())

    if (w1, w2, w3) in trigram_freq:
        return compute_trigram_probability(w1, w2, w3)
    elif (w2, w3) in bigram_freq:
        return alpha * (bigram_freq[(w2, w3)] / unigram_freq.get((w2, ), 1))
    else:
        return (alpha ** 2) * (unigram_freq.get((w3, ), 0) / total_unigrams)
```

### Test the Model

FInally let's create a function to predict the next word in a sentence.

```python
def predict_next_word(w1, w2):
    '''Predicts the next word using backoff probabilities.'''
    possible_words = [w[0] for w in unigram_freq.keys()]
    best_word = max(possible_words, key=lambda w3: backoff_probability(w1, w2, w3))
    return best_word
```

```python
# Example tokenization (you can use nltk or another tokenizer)
sentences = [
    'I realize that this may sound crazy, but I think I\'ve fallen in love with your younger',
    'The methodology used in his study is also helpful to us in conducting our',
    'I bought a house with the money that I inherited from my',
    'A man\'s happiness doesn\'t depend on what he has, but on what he'
]


for sentence in sentences:

    tokens = sentence.lower().split()  # Tokenize the sentence

    # Get the last two words of the sentence as context
    w1, w2 = tokens[-2], tokens[-1]

    # Predict the next word based on the context
    next_words = [predict_next_word(w1, w2) for _ in range(5)]

    # Print the predicted next word
    print(f'Sentence: {sentence}')
    print(f'Predicted 5 next word(s): {next_words} \n')
```

### Report Your Findings

All of the top 5 predicted words for each sentence end up being the same, which is good since our model is consistent. For the first sentence, we see the next predicted word is likely 'sister'. We can compare this with other words that the model could predict.

```python
w1 = 'your'
w2 = 'younger'
w3 = 'brother'
w4 = 'dog'
w5 = 'sister'

probs = [
    (w3, backoff_probability(w1, w2, w3)),
    (w4, backoff_probability(w1, w2, w4)),
    (w5, backoff_probability(w1, w2, w5))
    ]
print(sentences[0])
for word, prob in probs:
    print(f'Word: {word} \nProbability: {prob} \n')
```
